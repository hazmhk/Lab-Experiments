```python
import pandas as pd
data=pd.read_csv('/content/winequality-red.csv')
data
```

```python
import matplotlib.pyplot as plt
#create a Scatter plot of alcohol content vs pH

plt.scatter(data['alcohol'], data['pH'])
plt.xlabel('alcohol')
plt.ylabel('pH')
plt.show()
```

```python
import seaborn as sns
#create a pair plot of all columns in the dataset
sns.pairplot(data)
plt.show()
```

```python
#create a count plot of the quality

sns.countplot(x='quality', data=data)
plt.show()
```

```python
#create a heatmap of the correlation bw all columns of the dataset

sns.heatmap(data.corr(), annot=True, cmap='coolwarm') #false will hide values
plt.show()
```

```python
from wordcloud import WordCloud
#get column names of the dataset
column_names=data.columns.tolist()
#create a wordcloud plot of the column names
wordcloud=WordCloud(background_color='white').generate(''.join(column_names))
plt.imshow(wordcloud, interpolation = 'bilinear')#resie img
plt.axis('off')
plt.show()
```
