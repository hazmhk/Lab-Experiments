```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#read the dataset
data=pd.read_csv('/content/Salary_Data (1).csv')
data
data.dropna()
```

```python
#this displays the top 5 rows
data.head()
```

```python
#provides some info regarding the columns
data.info()
```

```python
#describes the basic state behind the dataset used
data.describe()
```

```python
x=data
x=data['Years of Experience']
x.head()
x.dropna()
```

```python
y=data
y=data['Salary']
y.head()
```

```python

from sklearn.model_selection import train_test_split
```

```python

x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.7, random_state=100)
```

```python

x_train=x_train[:,np.newaxis]
x_test=x_test[:,np.newaxis]
```

```python

from sklearn.linear_model import LinearRegression

```

```python

lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python

y_pred=lr.predict(x_test)
```

```python

c=[i for i in range(1, len(y_test)+1,1)]
plt.plot(c,y_test, color='r', linestyle='-')
plt.plot(c, y_pred, color='b', linestyle='-')
plt.xlabel('Salary')
plt.ylabel('index')
plt.title('prediction')
plt.show()
```

```python
#plotting the error
c=[i for i in range(1,len(y_test)+1,1)]
plt.plot(c, y_test-y_pred, color='green', linestyle='-')
plt.xlabel('index')
plt.ylabel('error')
plt.title('error value')
plt.show()
```

```python
#importng metrices for the evaluation of the model
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#calculte mean sq error
mse=mean_squared_error(y_test,y_pred)
#calc r sq value
rsq=r2_score(y_test,y_pred)

print(mse)
print(rsq)
```

```python
  #intercept and coeff of the line
print('intercept of the model:', lr.intercept_)
print('ccoeffcient of the line:', lr.coef_)
```
